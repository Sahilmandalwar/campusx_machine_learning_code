```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
from sklearn.datasets import make_classification
```

```python
X,y = make_classification(n_samples=100, n_features=2, n_redundant=0, n_informative=2, random_state=1, n_clusters_per_class=1)
plt.scatter(X[:, 0], X[:, 1], marker='o', c=y, cmap='winter')
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lor = LogisticRegression(solver='sag',penalty=None)
```

```python
lor.fit(X,y)
```

```python
m = -lor.coef_[0][0]/lor.coef_[0][1]
c = -lor.intercept_[0]/lor.coef_[0][1]
```

```python
plt.scatter(X[:, 0], X[:, 1], marker='o', c=y, cmap='winter')
x_point = np.linspace(-2,1)
plt.ylim(0,2.5)
plt.plot(x_point, m*x_point + c,color='green')
```

```python
def GradientDescent(X,y):
  X = np.insert(X,0,1,axis=1)
  weight = np.ones(X.shape[1])
  lr = 0.1
  epochs = 2000

  for epoch in range(epochs):
    y_hat = sigmoidFunction(np.dot(X,weight))
    weight = weight + lr*(np.dot((y-y_hat),X))/ X.shape[0]

  return weight[0] , weight[1:]
```

```python
def sigmoidFunction(n):
  return (1 + np.exp(-n))**(-1)
```

```python
intercept_ , coef_ = GradientDescent(X,y)
```

```python
m1 = -coef_[0] / coef_[1]
c1 = intercept_ / coef_[1]
```

```python
plt.scatter(X[:, 0], X[:, 1], marker='o', c=y, cmap='winter')
x_point = np.linspace(-2,1)
plt.ylim(0,2.5)
plt.plot(x_point, m*x_point + c,color='green')
plt.plot(x_point, m1*x_point + c1,color='red')
```
