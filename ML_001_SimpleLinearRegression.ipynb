```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/campusx-official/100-days-of-machine-learning/refs/heads/main/day48-simple-linear-regression/placement.csv")
```

```python
df.head()
```

```python
df.shape
```

```python
X = df['cgpa']
y = df['package']
```

```python
plt.scatter(df['cgpa'], df['package'], color='red')
plt.xlabel('cgpa')
plt.ylabel('package')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2 , random_state = 2)
```

```python
X_train
```

```python
y_train
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()  # linear regression object is created
```

```python
lr.fit(X_train.values.reshape(-1,1), y_train)  # object learn from data
```

```python
lr.predict(X_test.iloc[0].reshape(1,1))   # notice we are using .reshape() cause lr.predict require 2d object
```

```python
lr.predict(X_test.values.reshape(-1,1))  # predict whole data
```

```python
# find attributes
m = lr.coef_
c = lr.intercept_
```

```python
# required best fit line : y = m * x + c
m * X_test.iloc[0] + c
```

```python
x_points = np.linspace(4,10,100)  # random data for plotting line
```

```python
plt.scatter(df['cgpa'], df['package'], color='red')
plt.plot( x_points, m * x_points + c, color='blue')
plt.xlabel('cgpa')
plt.ylabel('package')
plt.show()   # visualise lr effect
```
