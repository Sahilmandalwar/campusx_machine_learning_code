```python
import numpy as np
from sklearn.datasets import make_classification
```

```python
X,y = make_classification(n_samples=100,n_classes=2,n_clusters_per_class=1,random_state=41,n_features=2,n_informative=1,n_redundant=0,class_sep=10,hypercube=False)
```

```python
import seaborn as sns
```

```python
sns.scatterplot(x=X[:,0],y=X[:,1],hue=y)
```

```python
def perceptron(X,y):
    X = np.insert(X,0,1,axis=1)
    weights = np.ones(X.shape[1])
    epoches = 1000
    lr = 0.1

    for i in range(epoches):
        j = np.random.randint(X.shape[0])
        y_hat = step(np.dot(weights,X[j]))
        weights = weights + lr*(y[j] - y_hat)*X[j]

    return weights
```

```python
def step(n):
  return 1 if n > 0 else 0
```

```python
C = perceptron(X,y)
```

```python
C
```

```python
m = -C[1]/C[2]
c = -C[0] / C[2]
```

```python
x_points = np.linspace(-3,3,100)
y_points = m*x_points + c
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(6,4))
plt.scatter(X[:,0],X[:,1],c=y)
plt.ylim(-4,3)
plt.plot(x_points,y_points)
```
