```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = 6 * np.random.rand(200,1) - 3
y = 0.9*X + 0.8*X**2 + 2 + np.random.rand(200, 1)*5
```

```python
plt.scatter(X,y)
plt.xlabel('X_value')
plt.ylabel("y_value")
plt.show()
```

```python
from sklearn.linear_model import LinearRegression, SGDRegressor
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
from sklearn.metrics import r2_score
from sklearn.pipeline import Pipeline
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=42, test_size=0.2)
```

```python
lr = LinearRegression()
lr.fit(X_train,y_train)
```

```python
lr.coef_, lr.intercept_
```

```python
plt.scatter(X,y,color='blue')
plt.xlabel('X_value')
plt.ylabel("y_value")
plt.plot(X_test,lr.coef_ * X_test + lr.intercept_,color='red')
plt.show()
```

```python
r2_score(y_test,lr.predict(X_test))
```

```python
poly = PolynomialFeatures(degree=2)
```

```python
X_train_data = poly.fit_transform(X_train)
```

```python
X_test_data = poly.transform(X_test)
```

```python
nlr = LinearRegression()
nlr.fit(X_train_data,y_train)
```

```python
nlr.intercept_, nlr.coef_
```

```python
r2_score(y_test,nlr.predict(X_test_data))
```

```python
x_new = np.linspace(-3,3,200).reshape(200,1)
print(x_new)
x_new_poly = poly.transform(x_new)
y_new = nlr.predict(x_new_poly)
```

```python
plt.scatter(X,y,color='blue')
plt.xlabel('X_value')
plt.ylabel("y_value")
plt.plot(x_new,y_new,color='red')
plt.show()
```

```python
def polyRegression(degrees):
  poly = PolynomialFeatures(degree=degrees)
  X_train_transformed = poly.fit_transform(X_train)

  X_test_transformed = poly.transform(X_test)

  lr = LinearRegression()
  lr.fit(X_train_transformed,y_train)

  x = np.linspace(-3,3,200).reshape(200,1)
  new_x = poly.transform(x)

  y_pred = lr.predict(new_x)
  plt.plot(x, y_pred, color='black')

  plt.scatter(X_test,y_test,color='red')
  plt.scatter(X_train,y_train,color='green')
  plt.show()


```
