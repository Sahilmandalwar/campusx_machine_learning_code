```python
import numpy as np
import pandas as pd
```

```python
from sklearn.datasets import make_classification
```

```python
X,y = make_classification(n_samples=10000,n_features=10,n_informative=3)
```

```python
X
```

```python
y
```

```python
from sklearn.ensemble import BaggingClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
```

```python
from sklearn.svm import SVC
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=42,test_size=0.2)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
clf = DecisionTreeClassifier(criterion='gini',max_depth=20)
```

```python
clf.fit(X_train,y_train)
```

```python
print("Accuracy Score: ", accuracy_score(y_test,clf.predict(X_test)))
```

Bagging : through decision tree

```python
bag = BaggingClassifier(estimator=DecisionTreeClassifier(),n_estimators=500,max_samples=0.25,bootstrap=True)
```

```python
bag.fit(X_train,y_train)
```

```python
print("Accuracy_score : ", accuracy_score(y_test,bag.predict(X_test)))
```

Pasting

```python
bag1 = BaggingClassifier(estimator=DecisionTreeClassifier(),n_estimators=500,max_samples=0.25,bootstrap=False)
bag1.fit(X_train,y_train)
print("Accuracy_score : ", accuracy_score(y_test,bag1.predict(X_test)))
```

```python
bag1.estimators_samples_
```

```python
bag1.estimators_features_
```

Random Subspaces

```python
bag2 = BaggingClassifier(estimator=DecisionTreeClassifier(),n_estimators=500,max_samples=1,max_features=0.5,bootstrap=True)
bag2.fit(X_train,y_train)
print("Accuracy_score : ", accuracy_score(y_test,bag2.predict(X_test)))
```

Random Patches

```python
bag3 = BaggingClassifier(estimator=DecisionTreeClassifier(),n_estimators=500,max_samples=0.38,max_features=0.7,bootstrap=True)
bag3.fit(X_train,y_train)
print("Accuracy_score : ", accuracy_score(y_test,bag3.predict(X_test)))
```

OOB score

```python
bag3 = BaggingClassifier(estimator=DecisionTreeClassifier(),n_estimators=500,max_samples=0.38,max_features=0.7,bootstrap=True,oob_score=True)
bag3.fit(X_train,y_train)
print("Accuracy_score : ", accuracy_score(y_test,bag3.predict(X_test)))
```

```python
bag3.oob_score_
```

Through Support Vector Classifier

```python
bag5 = BaggingClassifier(estimator=SVC(),n_estimators=50,max_samples=0.32,bootstrap=False)
bag5.fit(X_train,y_train)
print("Accuracy Score :", accuracy_score(y_test,bag5.predict(X_test)))
```
