```python
import pandas as pd
import numpy as np
import plotly.express as px
```

```python
from sklearn.datasets import make_regression
```

```python
X,y = make_regression(n_samples=100, n_features=2, n_informative=2, n_targets=1, noise=5)
```

```python
df = pd.DataFrame({'Feature_1' : X[:,0], "Feature_2" : X[:,1], "Target" : y})
```

```python
df.head()
```

```python
px.scatter_3d(df,x='Feature_1',y='Feature_2', z='Target')
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(df.iloc[:,:-1],df.iloc[:,-1],test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
y_predict = lr.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
print("MAE :", mean_absolute_error(y_predict, y_test))
```

```python
print("MSE :", mean_squared_error(y_predict, y_test))
```

```python
print("R2 Scored :", r2_score(y_predict, y_test))
```
